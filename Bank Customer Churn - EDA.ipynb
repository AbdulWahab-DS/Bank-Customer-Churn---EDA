## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```python
df = pd.read_csv('Customer-Churn-Records.csv')
```

## Data Understand

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Exited'].value_counts()
```

## Data Wrangling

```python
df.drop(['RowNumber','CustomerId','Surname'], axis=1, inplace = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
categorical_cols = ['Geography','Gender','HasCrCard','IsActiveMember','Complain']
for col in categorical_cols:
    df[col] = df[col].astype('category')
```

# Data Visualization

```python
ax = sns.countplot(x = 'Exited', data = df)
plt.title('Churn Distritbution')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
df['Exited'].value_counts(normalize = True) * 100
```

**“Majority of customers (around 80%) stayed with the bank, while 20% churned.”**

```python
ax = sns.countplot(x = 'Gender', hue = 'Exited', data = df)
plt.title('Gender vs Churn')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

**“Churn rate is slightly higher among females compared to males.”**

```python
ax = sns.countplot(x = 'Geography', hue = 'Exited', data = df)
plt.title('Geography vs Churn')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

**“Customers from Germany churn more than others.”**

```python
for col in ['HasCrCard','Card Type']:
    sns.countplot(x = col, hue = 'Exited', data = df)
```

```python
sns.countplot(x = 'IsActiveMember', hue = 'Exited' , data = df)
plt.title('Activity vs Churn')
plt.show()
```

**“Active members less likely to churn.”**

```python
ax = sns.countplot(x = 'Complain', hue = 'Exited', data = df)
plt.title('Complain vs Churn')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

**"Most customers who complained exited the bank, while the majority of non-complaining customers stayed."**

```python
sns.barplot(x = 'Exited', y = 'Age', data = df)
plt.title('Average Age vs Churn')
plt.show()
```

**“Older customers tend to churn more, whereas younger customers are more loyal.”**

```python
sns.barplot(x = 'Exited', y = 'Balance', data = df)
plt.title('Avg Balance vs Churn')
plt.show()
```

**“Churned customers have higher average balance than loyal customers.”**

```python
sns.barplot(x = 'Exited', y = 'Tenure', data = df)
plt.title('Avg Tenure vs Churn')
plt.show()
```

**“Long-term customers (high tenure) have lower churn rates.”**

```python
sns.barplot(x = 'Exited', y = 'EstimatedSalary', data = df)
plt.title('Avg EstimatedSalary vs Churn')
plt.show()
```

```python
sns.barplot(x = 'Exited', y = 'NumOfProducts', data = df)
plt.title('Avg Number of Products vs Churn')
plt.show
```

**“Customers with more products tend to stay loyal.”**

```python
sns.boxplot(x='Exited', y='CreditScore', data=df)
plt.title('Credit Score vs Churn')
plt.show()
```

```python
sns.boxplot(x = 'Exited', y = 'Point Earned', data = df)
plt.title('Point Earned vs Churn')
plt.show()
```

```python
sns.boxplot(x = 'Exited', y = 'Satisfaction Score', data = df)
plt.title('Satisfaction Score vs Churn')
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')
plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Numerical Features)')
plt.show()
```

# Conclusion

**“Overall, the majority of customers stayed with the bank, with a churn rate of around 20%. Churn is slightly higher among females and certain regions like Germany. Customer loyalty is influenced by activity and engagement—active members, long-term customers, and those with more products tend to remain, whereas older customers and those who filed complaints are more likely to churn. Interestingly, churned customers have higher average and median balances, indicating that balance alone does not guarantee retention.”**
